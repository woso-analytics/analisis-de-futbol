```python
path = r'C:\Users\MIRIRAI\Documents\Joel\Analisis-de-fulbo-main\WSL24-25' # use your path
all_files = glob.glob(os.path.join(path, "*.csv"))

df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
fotmob = pd.read_csv('fotmob.csv')
df = pd.merge(left=df, right=fotmob, left_on='teamName', right_on='club', how='left')
```

# Offside Trap

```python
# Function to calculate avg_x for a given team
def calculate_avg_x(team_name):
    dfteam = df[df['teamName'] == team_name]
    cb_df = dfteam[(dfteam['type'] == 'OffsideProvoked')]
    return cb_df['x'].median()

# Function to plot the defensive line for a team
def plot_def_line(ax, team_name):
    plt.rcParams['font.family'] = "DM Sans"
    dfteam = df[df['teamName'] == team_name]
    cb_df = dfteam[(dfteam['type'] == 'OffsideProvoked')]
    offpro = len(cb_df)
    avg_x = cb_df['x'].median()
    pitch = VerticalPitch(pitch_type='uefa', pitch_color='#EFE9E6', line_color=line_color, linewidth=1.5)
    pitch.draw(ax=ax)
    
    ax.hlines(avg_x, xmin=0, xmax=68, color=blue, linewidth=3, linestyle='--')
    # ax.hlines(cb_df.x, xmin=0, xmax=68, color=col2, linewidth=1, linestyle='--')
    ax.text(-10, avg_x, f'{avg_x:.1f}m', color=blue, fontsize=18, fontweight='bold', ha='center', va='center',
#             bbox=dict(facecolor=col1, edgecolor=blue, boxstyle='round,pad=0.3')
           )
    
    # ax.hlines(off_x, xmin=0, xmax=68, color=col1, linestyle='--', label='Offensive Actions')
    ax.text(34, 65, f'Offsides Provoked:\n{offpro}', color=blue, fontsize=22, fontweight='bold', ha='center', va='bottom',
            # bbox=dict(facecolor=col1, edgecolor=col1, boxstyle='round,pad=0.3')
           )
    
    # ax.hlines(def_x, xmin=0, xmax=68, color=col2, linestyle='--', label='Defensive Actions')
    ax.text(34, avg_x-3, f'Avg. Height of\nOffside Trap', color=bg_color, fontsize=15, fontweight='bold', ha='center', va='bottom',
            bbox=dict(facecolor=sky_blue, edgecolor=blue, boxstyle='round,pad=0.3'))

    fill_x = [0,68,68,0]
    fill_y = [0,0,avg_x,avg_x]
    ax.fill(fill_x, fill_y, color=sky_blue, alpha=0.5)
    
    ax.text(57, 114, team_name.replace('Women', ''), fontsize=18, fontweight='bold',  va='center', color=blue)
    fotmob_url = f'https://images.fotmob.com/image_resources/logo/teamlogo/{dfteam["fotmob_id"].iloc[0]}.png'
    image = Image.open(urllib.request.urlopen(f"{fotmob_url}"))
    ax_image = pitch.inset_image(114, 65, image, height=10, ax=ax)

    # ax.legend(loc='upper right')

fig,ax=plt.subplots(figsize=(10,10))
plot_def_line(ax, 'Manchester City Women')
```

```python
# List of team names
team_names = df.teamName.unique()

# Create a DataFrame with team names and their avg_x values
team_avg_x = pd.DataFrame({'teamName': team_names})
team_avg_x['avg_x'] = team_avg_x['teamName'].apply(calculate_avg_x)

# Sort the teams by avg_x value
team_avg_x_sorted = team_avg_x.sort_values(by='avg_x', ascending=False).dropna().reset_index(drop=True)

# Create the figure and axes for a 4x5 grid
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(18, 20), facecolor=bg_color)

# Flatten the axes array to easily loop through it
axes = axes.ravel()

# Loop through the sorted team names and corresponding axes
for i, row in enumerate(team_avg_x_sorted.itertuples()):
    team_name = row.teamName
    avg_x = row.avg_x
    plot_def_line(axes[i], team_name)
# plot_blank(axes[18])
# plot_blank(axes[11])

fig.text(0.21, 0.98, 'Barclays WSL Teams Offside Traps', fontsize=38, fontweight='bold', ha='left', va='top', color=blue)
fig.text(0.21, 0.955, "Manchester United have the highest offside trap and Manchester City/Aston Villa provoked the most offsides",
         fontsize=20, ha='left', va='top', color="#4E616C")
# fig_text(0.21, 0.94, "<Avg. Offensive Action Height of CenterBacks> | Overall Avg. Position of CneterBacks | <Avg. Defensive Action Height of CenterBacks>",
#          fontsize=17, ha='left', va='top', ax=fig, highlight_textprops=[{'color':col1}, {'color':col2}]_thin)
fig.text(0.21, 0.935, "Data from Opta | Barclays WSL 2024-25 ",
         fontsize=20, ha='left', va='top', color="#4E616C")

fotmob_url = 'https://images.fotmob.com/image_resources/logo/leaguelogo/9227.png'
image = Image.open(urllib.request.urlopen(f"{fotmob_url}"))
ax_image = add_image(image, fig, left=0.14, bottom=0.93, width=0.05, height=0.05)


fig_text(
    x = 0.75, y = 0.11,
    s = "@woso_analytics",
    color = '#1C2C5B',
    size = 12,
    annotationbbox_kw={"xycoords": "figure fraction"}
)

plt.savefig(
	"../figures/19052024_offsides_wsl.png",
	dpi = 600,
	facecolor = "#EFE9E6",
	bbox_inches="tight",
    edgecolor="none",
	transparent = False
)

plt.savefig(
	"../figures/19052024_offsides_wsl_tr.png",
	dpi = 600,
	facecolor = "none",
	bbox_inches="tight",
    edgecolor="none",
	transparent = True
)
```
