```python
green = '#69f900'
red = '#cb2217'
blue = '#00285e'
sky_blue = '#6caddf'
bg_color= '#fffcf7'
gold = '#ffc758'
line_color= '#000000'
# col1 = '#1C2C5B'
# col2 = '#6CABDD'
```

```python
path = r'C:\Users\MIRIRAI\Documents\Joel\Analisis-de-fulbo-main\WSL24-25' # use your path
all_files = glob.glob(os.path.join(path, "*.csv"))

df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
```

```python
def plot_blank(ax):
    pitch = VerticalPitch(pitch_type='uefa', pitch_color=bg_color, line_color=bg_color, corner_arcs=True)
    pitch.draw(ax=ax)
```

```python
def plot_goalkick(ax,team_name):
    plt.rcParams['font.family'] = "DM Sans"
    df_gk = df[(df['name']==team_name) & (df['qualifiers'].str.contains('GoalKick'))]
    pitch = VerticalPitch(pitch_type='uefa', pitch_color=bg_color, line_color='k', corner_arcs=True, line_zorder=5)
    pitch.draw(ax=ax)
    s=0
    u=0
    # setting the custom colormap
    pearl_earring_cmap1 = LinearSegmentedColormap.from_list("Pearl Earring - 10 colors", [bg_color, sky_blue], N=20)
    path_eff = [path_effects.Stroke(linewidth=3, foreground=bg_color), path_effects.Normal()]
    bin_statistic1 = pitch.bin_statistic(df_gk.endX, df_gk.endY, bins=(6,5), statistic='count', normalize=False)
    mask = bin_statistic1['statistic'] == 0
    bin_statistic1['statistic'] = np.ma.masked_where(mask, bin_statistic1['statistic'])
    pitch.heatmap(bin_statistic1, ax=ax, cmap=pearl_earring_cmap1, edgecolors=bg_color, linestyle='--', zorder=1)
    labels = pitch.label_heatmap(bin_statistic1, color='#202020', fontsize=20, ax=ax, zorder=5, ha='center', va='center', exclude_zeros=True, 
                                 str_format='{:.0f}', path_effects=path_eff)
    for index, row in df_gk.iterrows():
        if row['outcomeType']=='Successful':
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->', zorder=2, color=blue, #479bf5
                                            mutation_scale=7, alpha=0.85, linewidth=0.9)
            ax.add_patch(arrow)
            s+=1
        if row['outcomeType']=='Unsuccessful':
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->', zorder=2, color=red, 
                                            mutation_scale=7, alpha=0.85, linewidth=0.9)
            ax.add_patch(arrow)
            u+=1
    avg_dis = round(df_gk['endX'].mean(),2)
    ax.set_title(f'{team_name}',fontsize=22,fontweight='bold')
    ax_text(34, 107, s=f'Total: {s+u}, <Successful: {s}>, <Unsuccessful: {u}>',highlight_textprops=[{'color':blue},{'color':red}],
            ha='center',va='center',ax=ax,fontsize=16)
    ax.text(-2, avg_dis, f'Avg. Distance: {avg_dis}m',fontsize=16,rotation=90,ha='center',va='center')
    ax.hlines(avg_dis, xmin=0, xmax=68, color=gold, linestyle='--')
    # fig.text(0.25, 1.03, 'MCWFC Goal Kick Analysis', color=blue, fontweight='bold', fontsize=30, ha='left', va='center')
    # fig.text(0.25, 0.99, 'Barclays WSL | 24/25',color='#505050', fontsize=25, ha='left', va='center') 
    return

fig,axs=plt.subplots(1,2, figsize=(18,10), facecolor=bg_color)
plot_goalkick(axs[0],'Ayaka Yamashita')
plot_goalkick(axs[1],'Khiara Keating')

# fotmob_url = 'https://images.fotmob.com/image_resources/logo/leaguelogo/9227.png'
# club_icon = Image.open(urllib.request.urlopen(f'{fotmob_url}'))
# # image_ax.imshow(club_icon)
# # image_ax.axis('off')
# ax_timage = add_image(club_icon, fig, left=0.84, bottom=1, width=0.08, height=0.08)
fig_text(
    x = 0.8, y = .98,
    s = "| <Inside MCWFC>",
    highlight_textprops = [
        {"color": blue, "weight": "bold"}
    ],
    color = gold,
    size = 18,
    annotationbbox_kw={"xycoords": "figure fraction"}
)

fig_text(
    x = 0.75, y = 0.1,
    s = "Viz by: @joellikesthis",
    color = blue,
    size = 12,
    annotationbbox_kw={"xycoords": "figure fraction"}
)

# city_url = 'https://images.fotmob.com/image_resources/logo/teamlogo/231488.png'
# club_icon = Image.open(urllib.request.urlopen(f'{city_url}'))
# # image_ax.imshow(club_icon)
# # image_ax.axis('off')
# ax_timage = add_image(club_icon, fig, left=0.14, bottom=1, width=0.08, height=0.08)
```
