```python
def plot_blank(ax):
    pitch = VerticalPitch(pitch_type='uefa',pitch_color=bg_color,line_color='#f5f5f5',line_zorder=3)
    pitch.draw(ax=ax)
```

### Pass

```python
def pass_entry_for(ax,team_name):
    df_team = df[(df['teamName']==team_name) | (df['oppositionTeamName']==team_name)]
    df_team['next_qualifiers'] = df_team['qualifiers'].shift(-1)
    df_team['Length'] = np.sqrt((df['x']-df['endX'])**2 + (df['y']-df['endY'])**2)
    df_team = df_team[df_team['teamName']==team_name]
    
    box_entries = df_team[((df_team['type']=='Pass')|(df_team['type']=='Carry')) & (df_team['outcomeType']=='Successful') & (df_team['endX']>=88.5) &
                 ~((df_team['x']>=88.5) & (df_team['y']>=13.6) & (df_team['y']<=54.6)) & ((df_team['Length']>=5) | (df_team['Length']>=5)) &
                 (df_team['endY']>=13.6) & (df_team['endY']<=54.4) & (~df_team['qualifiers'].str.contains('CornerTaken|Freekick|ThrowIn'))]
    df_right = box_entries[(box_entries['type']=='Pass') & (box_entries['y']<=68/3)]
    fotmob_id = box_entries.fotmob_id.iloc[0]
    df_cent = box_entries[(box_entries['type']=='Pass') & (box_entries['y']>68/3) & (box_entries['y']<=136/3)]
    df_left = box_entries[(box_entries['type']=='Pass') & (box_entries['y']>136/3)]
    pitch = VerticalPitch(pitch_type='uefa',pitch_color=bg_color,line_color='k', line_zorder=3)
    pitch.draw(ax=ax)
    ax.set_ylim(52.25, 106)

    t=s=g=0
    for index, row in box_entries.iterrows():
        if (row['type']=='Pass'):
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->',
                                        color='k', zorder=4, mutation_scale=10, alpha=0.15, linewidth=1)
            t+=1
            ax.add_patch(arrow)
            ax.scatter(row['y'], row['x'], color='k', edgecolor='w', alpha=0.15, zorder=4)
        if (row['type']=='Pass' and ('KeyPass' in row['qualifiers'] or 'KeyPass' in row['next_qualifiers'])):
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->',
                                        color=blue, zorder=5, mutation_scale=10, alpha=0.6, linewidth=1.25)
            s+=1
            ax.add_patch(arrow)
            ax.scatter(row['y'], row['x'], color=blue, edgecolor='w', zorder=5)
        if (row['type']=='Pass' and ('IntentionalGoalAssist' in row['qualifiers'] or 'IntentionalGoalAssist' in row['next_qualifiers'])):
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->',
                                        color=red, zorder=6, mutation_scale=15, alpha=1, linewidth=2)
            g+=1
            ax.add_patch(arrow)
            ax.scatter(row['y'], row['x'], color=red, edgecolor='w', zorder=6)

    left_counts = df_left['shortName'].value_counts()
    max_player_left = left_counts.idxmax()
    max_count_left = left_counts.max()

    cent_counts = df_cent['shortName'].value_counts()
    max_player_cent = cent_counts.idxmax()
    max_count_cent = cent_counts.max()

    right_counts = df_right['shortName'].value_counts()
    max_player_right = right_counts.idxmax()
    max_count_right = right_counts.max()
    
    
    # Separate left from right hand side
    # -- Count which side had most crosses:
    left_side = df_left.shape[0]
    right_side = df_right.shape[0]
#     left_success = other_s[other_s['y'] > 50].shape[0]
#     right_success = other_s[other_s['y'] < 50].shape[0]

    if right_side > left_side:
        color_left = '#F5F1E3'
        color_right = '#DDDBCB'
    else:
        color_right = '#F5F1E3'
        color_left = '#DDDBCB'

    ax.fill_between(
        x=[0.35,33.3], y1=ax.get_ylim()[0], y2=104.65,
        color=color_right, zorder=-1
    )
    ax.fill_between(
        x=[68,33.3], y1=ax.get_ylim()[0], y2=104.65,
        color=color_left, zorder=-1
    )


    fig_text(
    x = 0.445, y = .86, 
    s = f"{team_name}",
    va = "bottom", ha = "center",
    fontsize = 20, color =blue , weight = "bold"
    )
    fig_text(
        x = 0.26, y = .83,
        s=f'Box entries by open-play Passes: {t}, <Led to shot: {s}>, <Led to goal: {g}>', fontsize=12, color='#4E616C', ax=ax,
            highlight_textprops=[{'color':blue}, {'color':red}], 
        ha='left', 
        va='center',
    )


    ax.text(70, 45, f'''
    Most From Left: {max_player_left} ({max_count_left})  |  Most From Center: {max_player_cent} ({max_count_cent}) |   Most From Right: {max_player_right} ({max_count_right})
    Entries From Left: {len(df_left)} |  Entries From Centre: {len(df_cent)}  | Entries From The Right: {len(df_right)}
    ''', 
    fontsize=12
           )
    
    fig_text(
        x = 0.78, y = .79,
        s = "| <Inside MCWFC>",
        highlight_textprops = [
            {"color": blue, "weight": "bold"}
        ],
        color = gold,
        size = 12,
        annotationbbox_kw={"xycoords": "figure fraction"}
    )

    
    
    # Take data coordinates and transform them to normalized figure coordinates
    fotmob_url = 'https://images.fotmob.com/image_resources/logo/leaguelogo/9227.png'
    club_icon = Image.open(urllib.request.urlopen(f'{fotmob_url}'))
    # image_ax.imshow(club_icon)
    # image_ax.axis('off')
    ax_timage = add_image(club_icon, fig, left=0.84, bottom=0.85, width=0.040, height=0.050)

    city_url = f'https://images.fotmob.com/image_resources/logo/teamlogo/{fotmob_id}.png'
    club_icon = Image.open(urllib.request.urlopen(f'{city_url}'))
    # image_ax.imshow(club_icon)
    # image_ax.axis('off')
    ax_timage = add_image(club_icon, fig, left=0.14, bottom=0.85, width=0.050, height=0.050)
    
    return

fig,ax=plt.subplots(figsize=(10,10),facecolor=bg_color)
plt.rcParams['font.family'] = "DM Sans"
pass_entry_for(ax,'Manchester City Women')
```

```python
def carry_entry_for(ax,team_name):
    df_team = df[(df['teamName']==team_name) | (df['oppositionTeamName']==team_name)]
    df_team['next_qualifiers'] = df_team['qualifiers'].shift(-1)
    df_team['Length'] = np.sqrt((df['x']-df['endX'])**2 + (df['y']-df['endY'])**2)
    df_team = df_team[df_team['teamName']==team_name]
    
    box_entries = df_team[((df_team['type']=='Pass')|(df_team['type']=='Carry')) & (df_team['outcomeType']=='Successful') & (df_team['endX']>=88.5) &
                 ~((df_team['x']>=88.5) & (df_team['y']>=13.6) & (df_team['y']<=54.6)) & ((df_team['Length']>=5) | (df_team['Length']>=5)) &
                 (df_team['endY']>=13.6) & (df_team['endY']<=54.4) & (~df_team['qualifiers'].str.contains('Corner|Freekick|Cross|ThrowIn'))]
    fotmob_id = box_entries.fotmob_id.iloc[0]
    df_right = box_entries[(box_entries['type']=='Carry') & (box_entries['y']<=68/3)]
    df_cent = box_entries[(box_entries['type']=='Carry') & (box_entries['y']>68/3) & (box_entries['y']<=136/3)]
    df_left = box_entries[(box_entries['type']=='Carry') & (box_entries['y']>136/3)]
    pitch = VerticalPitch(pitch_type='uefa',pitch_color=bg_color,line_color='k',corner_arcs=True,line_zorder=3)
    pitch.draw(ax=ax)
    ax.set_ylim(52.25, 106)

    t=s=g=0
    for index, row in box_entries.iterrows():
        if row['type']=='Carry':
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->', linestyle='--',
                                        color='k', zorder=4, mutation_scale=10, alpha=0.25, linewidth=1)
            t+=1
            ax.add_patch(arrow)
            ax.scatter(row['y'], row['x'], color='k', edgecolor='w', alpha=0.25, zorder=4)
        if (row['type']=='Carry' and ('KeyPass' in row['qualifiers'] or 'KeyPass' in row['next_qualifiers'])):
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->', linestyle='--',
                                        color=col2, zorder=5, mutation_scale=10, alpha=1, linewidth=1.25)
            s+=1
            ax.add_patch(arrow)
            ax.scatter(row['y'], row['x'], color=col2, edgecolor='w', zorder=5)
        if (row['type']=='Carry' and ('IntentionalGoalAssist' in row['qualifiers'] or 'IntentionalGoalAssist' in row['next_qualifiers'])):
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->', linestyle='--',
                                        color='green', zorder=6, mutation_scale=15, alpha=1, linewidth=2)
            g+=1
            ax.add_patch(arrow)
            ax.scatter(row['y'], row['x'], color='green', edgecolor='w', zorder=6)

    if df_left.empty:
        max_player_left = 'None'
        max_count_left = 0
    else:   
        left_counts = df_left['shortName'].value_counts()
        max_player_left = left_counts.idxmax()
        max_count_left = left_counts.max()
#     ax.text(68, 47, f"Most from left:\n{max_player_left}:{max_count_left}", color='k', fontsize=18, va='center', ha='left')
    
    if df_cent.empty:
        max_player_cent = 'None'
        max_count_cent = 0
    else:   
        cent_counts = df_cent['shortName'].value_counts()
        max_player_cent = cent_counts.idxmax()
        max_count_cent = cent_counts.max()
#     ax.text(34, 47, f"Most from center:\n{max_player_cent}:{max_count_cent}", color='k', fontsize=18, va='center', ha='center')
    
    if df_right.empty:
        max_player_rigt = 'None'
        max_count_rigt = 0
    else:   
        right_counts = df_right['shortName'].value_counts()
        max_player_right = right_counts.idxmax()
        max_count_right = right_counts.max()
#     ax.text(0, 47, f"Most from right:\n{max_player_rigt}:{max_count_rigt}", color='k', fontsize=18, va='center', ha='right')

    # Separate left from right hand side
    # -- Count which side had most crosses:
    left_side = df_left.shape[0]
    right_side = df_right.shape[0]
#     left_success = other_s[other_s['y'] > 50].shape[0]
#     right_success = other_s[other_s['y'] < 50].shape[0]

    if right_side > left_side:
        color_left = '#F5F1E3'
        color_right = '#DDDBCB'
    else:
        color_right = '#F5F1E3'
        color_left = '#DDDBCB'

    ax.fill_between(
        x=[0.35,34.2], y1=ax.get_ylim()[0], y2=104.65,
        color=color_right, zorder=-1
    )
    ax.fill_between(
        x=[68,34.2], y1=ax.get_ylim()[0], y2=104.65,
        color=color_left, zorder=-1
    )


    fig_text(
    x = 0.445, y = .86, 
    s = f"{team_name}",
    va = "bottom", ha = "center",
    fontsize = 20, color =blue , weight = "bold"
    )
    fig_text(
        x = 0.26, y = .83,
        s=f'Box entries by Carries: {t}, <Led to shot: {s}>, <Led to goal: {g}>', fontsize=12, color='#4E616C', ax=ax,
            highlight_textprops=[{'color':blue}, {'color':red}], 
        ha='left', 
        va='center',
    )


    ax.text(70, 45, f'''
    Most From Left: {max_player_left} ({max_count_left})  |  Most From Center: {max_player_cent} ({max_count_cent}) |   Most From Right: {max_player_right} ({max_count_right})
    Entries From Left: {len(df_left)} |  Entries From Centre: {len(df_cent)}  | Entries From The Right: {len(df_right)}
    ''', 
    fontsize=12
           )
    
    fig_text(
        x = 0.78, y = .79,
        s = "| <Inside MCWFC>",
        highlight_textprops = [
            {"color": blue, "weight": "bold"}
        ],
        color = gold,
        size = 12,
        annotationbbox_kw={"xycoords": "figure fraction"}
    )

    
    
    # Take data coordinates and transform them to normalized figure coordinates
    fotmob_url = 'https://images.fotmob.com/image_resources/logo/leaguelogo/9227.png'
    club_icon = Image.open(urllib.request.urlopen(f'{fotmob_url}'))
    # image_ax.imshow(club_icon)
    # image_ax.axis('off')
    ax_timage = add_image(club_icon, fig, left=0.84, bottom=0.85, width=0.040, height=0.050)

    city_url = f'https://images.fotmob.com/image_resources/logo/teamlogo/{fotmob_id}.png'
    club_icon = Image.open(urllib.request.urlopen(f'{city_url}'))
    # image_ax.imshow(club_icon)
    # image_ax.axis('off')
    ax_timage = add_image(club_icon, fig, left=0.14, bottom=0.85, width=0.050, height=0.050)
    
    return

fig,ax=plt.subplots(figsize=(10,10),facecolor=bg_color)
carry_entry_for(ax,'Manchester City Women')
```
