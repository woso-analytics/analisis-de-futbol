```python
df = pd.read_csv('../data/contracts.csv')
```

```python
df
```

```python
# df = df.loc[~df['in_the_team'].str.contains('?')]
df = df.loc[df['in_the_team'] != '?']
df['in_the_team'] = df['in_the_team'].str.replace('-?', '-unkwown', regex=False)
df['in_the_team'] = df['in_the_team'].str.replace(" ", "")
df['in_the_team'] = df['in_the_team'].str.replace('?', '', regex=False)
```

```python
df['start_year'] = df['in_the_team'].apply(lambda x: x.split('-')[0])
df['end_year'] = df['in_the_team'].apply(lambda x: x.split('-')[1] if len(x.split('-'))>1 else 'unknown')
```

```python
df['position']= df['name_position'].apply(lambda x: x.split()[-1])
df['name']= df['name_position'].apply(lambda x: x.rsplit(' ', 1)[0])
```

```python
# Function to extract short names
def get_short_name(full_name):
    if pd.isna(full_name):
        return full_name
    parts = full_name.split()
    if len(parts) == 1:
        return full_name  # No need for short name if there's only one word
    elif len(parts) == 2:
        return parts[0][0] + ". " + parts[1]
    else:
        return parts[0][0] + ". " + parts[1][0] + ". " + " ".join(parts[2:])

# Applying the function to create 'shortName' column
df['name'] = df['name'].apply(get_short_name)
```

```python
df = df.drop(columns=['shirtno', 'name_position', 'foot', 'height', 'nat', 'in_the_team', 'start_year'])
```

```python
data_clean = df.loc[(df['team'] == 'Leicester City') & (df['end_year'] == '2025')]
data_clean
```

```python
def plot_table(ax, teamName):
    data_clean = df.loc[(df['team'] == teamName) & (df['end_year'] == '2025')]
    
    nrows = data_clean.shape[0]
    ncols = data_clean.shape[1] - 3 # because I want to remove aprox. 3 columns from my DF
    
    ax.set_xlim(0, 6)
    ax.set_ylim(-3, 5)
    
    # calculate asymmetry of x and y axes: (I have no idea how this works)
    x0, y0 = ax.transAxes.transform((0, 0)) # lower left in pixels
    x1, y1 = ax.transAxes.transform((1, 1)) # upper right in pixes
    dx = x1 - x0
    dy = y1 - y0
    maxd = max(dx, dy)
    width = .26 * maxd / dx
    height = .81 * maxd / dy
    
    # Iterate
    for y in range(0, nrows):
        # col = '#009FB7'
        if data_clean['position'].iloc[y] =='Goalkeeper':
            col = '#009FB7'
        elif data_clean['position'].iloc[y] =='Defence':
            col = '#003049'
        elif data_clean['position'].iloc[y] =='Midfield':
            col = '#F77F00'
        else:
            col = '#96031A'
        
        # -- Player name
        ax_text(
            x=0.5, y=4- .5*y,
            s=data_clean['name'].iloc[y],
            fontfamily='DM Sans',
            size=8,
            ha='left', va='center', ax=ax
        )
        
        # -- Age
        ax_text(
            x=5, y=4- .5*y,
            s=f"{data_clean['age'].iloc[y]}",
            fontfamily='DM Sans',
            size=8,
            ha='center', va='center', ax=ax
        )
        ax.scatter(
                x=3.5,
                y=4-.5*y,
                s=30,
                color=col,
                marker='s',
                edgecolor='#00285E',
                linewidth=.8
            )
    #team name
    ax.text(2, 5.5, f'{teamName}' , fontsize=9, fontweight='bold',  ha='center', va='center', color=blue)
    ax.set_axis_off()
    
    
    # -- Column titles
    ax_text(
        x=5, y=5,
        s='Age',
        size=8,
        ha='center', va='center', ax=ax,
        textalign='center', weight='bold'
    )
```

```python
# Create the main figure
fig = plt.figure(figsize=(8,14))
plt.rcParams['font.family'] = 'DM Sans'

# --------------- Header Section ---------------
ax1 = fig.add_axes([0, 0.7, .75, .15])
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)

# Header Text
ax1.text(0.5, 0.93, 'Barclays WSL: Expiring Contracts (inc loans)', fontfamily='Cal Sans', fontsize=16, fontweight='bold', color='#00285E', ha='center')
ax1.text(0.5, 0.78, 'All Players whose contracts expire at the end of the season', fontsize=10, fontweight='bold', color='#00285E', ha='center')

# Position Titles
positions = ['Goalkeeper', 'Defender', 'Midfielder', 'Forward']
x_positions = [0.20, 0.39, 0.58, 0.76]
for i, pos in enumerate(positions):
    ax1.text(x_positions[i], 0.61, pos, fontsize=8, color='#00285E', ha='center')

# Add scatter points for positions
colors = ['#009FB7', '#003049', '#F77F00', '#96031A']
color_positions = [0.26, 0.44, 0.635, 0.81]
for x, color in zip(color_positions, colors):
    ax1.scatter(x, 0.63, s=50, color=color, marker='s', edgecolor='#00285E', linewidth=0.8)

ax1.set_axis_off()  # Hide the header axis

# --------------- Plot Section ---------------
# Main table plot
ax2 = fig.add_axes([0.05, 0.25, 0.75, .5])  # x0, y0, width, height
ax2.set_axis_off()
teams = df.team.unique()
for i, team in enumerate(teams):
    row = i // 3
    col = i % 3
    subplot = ax2.inset_axes([col * 0.3, 1 - (row + 1) * 0.25, 0.35, 0.25])
    plot_table(subplot, team)


# Save the figure with both background and transparent options
plt.savefig("../figures/wsl_free_transfers.png", dpi=600, facecolor="#EFE9E6", bbox_inches="tight", transparent=False)
plt.savefig("../figures/wsl_free_transfers_tr.png", dpi=600, facecolor="none", bbox_inches="tight", transparent=True)
```
