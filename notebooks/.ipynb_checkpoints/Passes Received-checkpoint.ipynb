```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.markers import MarkerStyle
from mplsoccer.pitch import Pitch
from mplsoccer import Pitch, VerticalPitch
import matplotlib.patches as patches
from collections import Counter
import matplotlib.patheffects as path_effects
import glob
import os

import matplotlib.cm as cm
import matplotlib.colors as mcolors

from highlight_text import ax_text, fig_text
from PIL import Image
from mplsoccer import add_image
import urllib
from urllib.request import urlopen
```

```python
path = r'C:\Users\Hp2\Music\joel\An√°lisis de fulbo\WSL24-25' # use the path with your data
all_files = glob.glob(os.path.join(path, "*.csv"))

df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
```

```python
fotmob = pd.read_csv('fotmob.csv')
df = pd.merge(left=df, right=fotmob, left_on='teamName', right_on='club', how='left')
```

```python
df = df[df['type']!='Carry'].reset_index(drop=True)
```

```python
green = '#69f900'
red = '#cb2217'
blue = '#00285e'
sky_blue = '#6caddf'
bg_color= '#fffcf7'
gold = '#ffc758'
line_color= '#000000'
bbox_pad = .8
bboxprops = {'linewidth': 0, 'pad': bbox_pad}
```

```python
def plot_pass_recieve(ax, pname):
    dfp = df[(df['type']=='Pass') & (df['outcomeType']=='Successful') & (df['name'].shift(-1)==pname)]
    dfkp = dfp[dfp['qualifiers'].str.contains('KeyPass')]
    dfas = dfp[dfp['qualifiers'].str.contains('IntentionalGoalAssist')]
    dfnt = dfp[dfp['endX']>=70]
    dfpen = dfp[(dfp['endX']>=87.5) & (dfp['endY']>=13.6) & (dfp['endY']<=54.6)]
    dfpro = dfp[(dfp['x']>=35) & (dfp['prog_pass']>=9.11)]
    dfcros = dfp[(dfp['qualifiers'].str.contains('Cross'))]
    dfxT = dfp[dfp['xT']>=0]
    dflb = dfp[(dfp['qualifiers'].str.contains('Longball')) & (dfp['x']<=52)]
    cutback = dfp[((dfp['x'] >= 88.54) & (dfp['x'] <= 105) & 
                       ((dfp['y'] >= 40.8) & (dfp['y'] <= 54.4) | (dfp['y'] >= 13.6) & (dfp['y'] <= 27.2)) & 
                       (dfp['endY'] >= 27.2) & (dfp['endY'] <= 40.8) & (dfp['endX'] >= 81.67))]

    name_counts = dfp['shortName'].value_counts()
    name_counts_df = name_counts.reset_index()
    name_counts_df.columns = ['name', 'count']
    name_counts_df = name_counts_df.sort_values(by='count', ascending=False)  
    name_counts_df = name_counts_df.reset_index()
    fotmob_id = dfp.fotmob_id.iloc[0]

    pitch = Pitch(pitch_type='uefa', pitch_color=bg_color, line_color='k', linewidth=2)
    pitch.draw(ax=ax)
    # pr = pitch.lines(dfp.x, dfp.y, dfp.endX, dfp.endY, color='#2680ff', lw=2.25, alpha=0.25, comet=True, transparent=True, ax=ax)
    kp = pitch.lines(dfkp.x, dfkp.y, dfkp.endX, dfkp.endY, color=blue, lw=2.35, comet=True, transparent=True, ax=ax)
    ass = pitch.lines(dfas.x, dfas.y, dfas.endX, dfas.endY, color=red, lw=2.5, comet=True, transparent=True, ax=ax)

    ax.scatter(dfp.endX, dfp.endY, s=35, color=bg_color, alpha=0.5, edgecolor=sky_blue, zorder=1)
    ax.scatter(dfkp.endX, dfkp.endY, s=35, color=bg_color, edgecolor=blue, zorder=3)
    ax.scatter(dfas.endX, dfas.endY, s=35, color=bg_color, edgecolor=red, zorder=3)

    ax.hlines(dfp['endY'].median(), xmin=-2, xmax=107, color='gray', linewidth=2, linestyle='--')
    ax.vlines(dfp['endX'].median(), ymin=-2, ymax=70, color='gray', linewidth=2, linestyle='--')

    plt.rcParams['font.family'] = "DM Sans"
    fig_text(
    x = 0.445, y = .81, 
    s = f"{pname} Passes Received",
    va = "bottom", ha = "center",
    fontsize = 20, color =blue , weight = "bold"
    )
    
    fig_text(
        x = 0.46, y = .78, 
        s='<Passes Received> | <Key Passes Received> | <Assists Received> ',
        highlight_textprops=[{'color': sky_blue},{'color':blue},{'color': red}],
        va = "bottom", ha = "center",
        fontsize = 12, color = '#4E616C'
    )
#     str_text = f'''
#     <Passes Received> | <Key Passes Received> | <Assists Received> 
#     '''
#     fig_text(
# 	x = 0.2, y = .81, 
#     s = str_text,
#     highlight_textprops=[{'color': 'w', 'weight': 'bold', 'bbox': {'facecolor':sky_blue, **bboxprops}},
#                         {'color': 'w', 'weight': 'bold', 'bbox': {'facecolor':blue, **bboxprops}},
#                         {'color': 'w', 'weight': 'bold', 'bbox': {'facecolor':red, **bboxprops}},
#                         ],)

    fig_text(
        x = 0.8, y = .76,
        s = "| <Inside MCWFC>",
        highlight_textprops = [
            {"color": blue, "weight": "bold"}
        ],
        color = gold,
        size = 12,
        annotationbbox_kw={"xycoords": "figure fraction"}
    )


    ax.text(1, .4, f'''
    Passes Received: {len(dfp)}  |  Key Passes Received: {len(dfkp)} |   Assists Received: {len(dfas)}
    Final 1/3 Passes Received: {len(dfnt)} |  Passes Received in Pen. box: {len(dfpen)}  | Progressive Passes Received: {len(dfpro)}
    Crosses Received: {len(dfcros)} | Longballs Received from own half: {len(dflb)}  |  xT Received by Pass: {dfxT['xT'].sum().round(2)}
    Avg. reception distance from opponent Goal line: {round(105-dfp['endX'].median(),2)}m  |  Most Passes from: {name_counts_df['name'][0]} ({name_counts_df['count'][0]})''', fontsize=10, ha='left', va='top')

    team_name = dfp['teamName'].unique()
    team_name = team_name[0]
    # Take data coordinates and transform them to normalized figure coordinates
    fotmob_url = 'https://images.fotmob.com/image_resources/logo/leaguelogo/9227.png'
    club_icon = Image.open(urllib.request.urlopen(f'{fotmob_url}'))
    # image_ax.imshow(club_icon)
    # image_ax.axis('off')
    ax_timage = add_image(club_icon, fig, left=0.84, bottom=0.8, width=0.040, height=0.050)

    city_url = f'https://images.fotmob.com/image_resources/logo/teamlogo/{fotmob_id}.png'
    club_icon = Image.open(urllib.request.urlopen(f'{city_url}'))
    # image_ax.imshow(club_icon)
    # image_ax.axis('off')
    ax_timage = add_image(club_icon, fig, left=0.14, bottom=0.8, width=0.050, height=0.050)
    
    return

fig,ax=plt.subplots(figsize=(10,10), facecolor=bg_color)
plot_pass_recieve(ax, 'Vivianne Miedema')
```
