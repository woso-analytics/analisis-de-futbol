```python
def calculate_avg_h(df, team_name):
    team_df = df[(df['teamName']==team_name) & (~df['qualifiers'].str.contains('FreeKick|Corner|ThrowIn'))]
    right = team_df[team_df['position'].isin(['DR', 'DMR'])]
    left = team_df[team_df['position'].isin(['DL', 'DML'])]
    
#     rw = team_df[team_df['position'].isin(['FWR'])] else team_df[team_df['position'].isin(['AML'])]
#     lw = team_df[team_df['position'].isin(['FWL'])] else  team_df[team_df['position'].isin(['AML'])]
    
    avg_hr = right['x'].median()
    avg_hl = left['x'].median()
    avg_h = (avg_hr + avg_hl) / 2
    return avg_h

# def plot_fullback(ax, df, team_name, image, avg_h):
def plot_fullback(ax, df, team_name, avg_h):
    team_df = df[(df['teamName']==team_name) & (~df['qualifiers'].str.contains('FreeKick|Corner|ThrowIn|BallRecovery|Defensive|Interception|Tackle|Aerial|Clearance|BlockedPass|Challenge'))]
    right = team_df[team_df['position'].isin(['DR', 'DMR'])]
    left = team_df[team_df['position'].isin(['DL', 'DML'])]
    
    
    rw = team_df[team_df['position'].isin(['FWR'])] 
    lw = team_df[team_df['position'].isin(['FWL'])]
    
    if len(rw)<1:
         rw =  team_df[team_df['position'].isin(['AMR'])]
    if len(lw)<1:
         lw = team_df[team_df['position'].isin(['AML'])]
    
    #FBS
    avg_hr = right['x'].median()
    avg_hl = left['x'].median()
    avg_wr = right['y'].median()
    avg_wl = left['y'].median()
    dist = avg_wl - avg_wr
    
    #WIngers
    avg_hrw = rw['x'].median()
    avg_hlw = lw['x'].median()
    avg_wrw = rw['y'].median()
    avg_wlw = lw['y'].median()
    dist = avg_wlw - avg_wrw

    pitch = VerticalPitch(pitch_type='uefa', pitch_color=bg_color, line_color='gray', linewidth=2)
    pitch.draw(ax=ax)
    
    # Plot lines and scatter points for FB positions
    pitch.lines(avg_hr, avg_wr, avg_hl, avg_wl, color=col1, lw=7, ax=ax)
    pitch.scatter(avg_hr, avg_wr, s=1000, color=col1, edgecolor=bg_color, lw=3, zorder=3, ax=ax)
    pitch.scatter(avg_hl, avg_wl, s=1000, color=col1, edgecolor=bg_color, lw=3, zorder=3, ax=ax)

     # Plot lines and scatter points for winger positions
    pitch.lines(avg_hrw, avg_wrw, avg_hlw, avg_wlw, color=blue, lw=7, ax=ax)
    pitch.scatter(avg_hrw, avg_wrw, s=1000, color=blue, edgecolor=bg_color, lw=3, zorder=3, ax=ax)
    pitch.scatter(avg_hlw, avg_wlw, s=1000, color=blue, edgecolor=bg_color, lw=3, zorder=3, ax=ax)
    
    # Annotate the plot
    #FBs
    ax.text(avg_wr, avg_hr+2, 'RB', fontsize=20, ha='center', va='top', color='white')
    ax.text(avg_wl, avg_hl+2, 'LB', fontsize=20, ha='center', va='top', color='white')
    ax.text(avg_wr, avg_hr-13, f'{avg_hr:.1f}', fontsize=20, ha='center', va='bottom')
    ax.text(avg_wl, avg_hl-13, f'{avg_hl:.1f}', fontsize=20, ha='center', va='bottom')
    
    #Wingers
    ax.text(avg_wrw, avg_hrw+2, 'RW', fontsize=19, ha='center', va='top', color='white')
    ax.text(avg_wlw, avg_hlw+2, 'LW', fontsize=19, ha='center', va='top', color='white')
    ax.text(avg_wrw, avg_hrw+6, f'{avg_hr:.1f}', fontsize=20, ha='center', va='bottom')
    ax.text(avg_wlw, avg_hlw+6, f'{avg_hl:.1f}', fontsize=20, ha='center', va='bottom')
    
    
#     ax.text(34, 114, f'Avg. Height: {avg_h:.1f}m', fontsize=13, ha='center', va='bottom')
#     ax.text(34, 107, f'Avg. Width: {dist:.1f}m', fontsize=13, ha='center', va='bottom')
    ax.text(34, 116, team_name, fontsize=23, fontweight='bold', ha='center', va='bottom')

    # Insert preloaded image
#     pitch.inset_image(52.5, 34, image, height=25, alpha=0.5, zorder=1, ax=ax)

team_names = df.teamName.unique()

# # Preload team logos
# team_images = {}
# for team_name in team_names:
#     IMAGE_URL = f'file:///D:/FData/club_teams_logo/all_league%20-%20Copy/{team_name}.html'
#     team_images[team_name] = Image.open(urlopen(IMAGE_URL))

# Create a DataFrame with team names and their avg_h values
team_avg_h = pd.DataFrame({'teamName': team_names})
team_avg_h['avg_h'] = team_avg_h['teamName'].apply(lambda x: calculate_avg_h(df, x))
team_avg_h_sorted = team_avg_h.sort_values(by='avg_h', ascending=False).reset_index(drop=True)

# Create the figure and axes for a 4x5 grid
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(24, 18), facecolor=bg_color)

# Flatten the axes array to easily loop through it
axes = axes.ravel()

# Loop through the sorted team names and corresponding axes
for i, row in enumerate(team_avg_h_sorted.itertuples()):
    team_name = row.teamName
    avg_h = row.avg_h
    plot_fullback(axes[i], df, team_name, avg_h)


fig.text(0.22, 1.01, "Avg. Height of the Wide Players",
         fontsize=32, ha='left', va='top', color=blue)
fig.text(0.22, 0.98, "GW21 | 24-25",
         fontsize=24, ha='left', va='top', color = "#373737")

fotmob_url = 'https://images.fotmob.com/image_resources/logo/leaguelogo/9227.png'
image = Image.open(urllib.request.urlopen(f"{fotmob_url}"))
ax_image = add_image(image, fig, left=0.14, bottom=.97, width=0.06, height=0.06)
```
