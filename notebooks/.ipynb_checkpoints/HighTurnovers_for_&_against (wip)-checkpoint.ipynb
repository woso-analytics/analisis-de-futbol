```python
def team_ht_list(team_name):
    df_team = df[(df['teamName']==team_name) | (df['oppositionTeamName']==team_name)]
    df_team = df_team.reset_index(drop=True)
    
    ht_count = 0
    # Iterate through the DataFrame
    for i in range(len(df_team)):
        if ((df_team.loc[i, 'type'] in ['BallRecovery', 'Interception']) and 
            (df_team.loc[i, 'teamName'] == team_name) and 
            (df_team.loc[i, 'x'] >= 63)):
            
            # Check the following rows
            j = i + 1
            if ((df_team.loc[j, 'teamName']==team_name) and
                (df_team.loc[j, 'type']!='Dispossessed') and (df_team.loc[j, 'type']!='OffsidePass')):
                ax.scatter(df_team.loc[i, 'y'],df_team.loc[i, 'x'], s=55, color='None', edgecolor='#ff4b44')
                ht_count += 1

    return ht_count

def plot_highTurnover(ax, team_name):
    df_team = df[(df['teamName']==team_name) | (df['oppositionTeamName']==team_name)]
    df_team = df_team.reset_index(drop=True)

    pitch = VerticalPitch(pitch_type='uefa', pitch_color='#f5f5f5', line_color='k', linewidth=2, corner_arcs=True)
    pitch.draw(ax=ax)
    ax.set_ylim(52, 105.5)
    # ax.set_xlim(68.5 , -0.5)

    goal_count = 0
    # Iterate through the DataFrame
    for i in range(len(df_team)):
        if ((df_team.loc[i, 'type'] in ['BallRecovery', 'Interception']) and 
            (df_team.loc[i, 'teamName'] == team_name) and 
            (df_team.loc[i, 'x'] >= 63)):
            
            possession_id = df_team.loc[i, 'possession_id']
            
            # Check the following rows within the same possession
            j = i + 1
            while j < len(df_team) and df_team.loc[j, 'possession_id'] == possession_id and df_team.loc[j, 'teamName']==team_name:
                if df_team.loc[j, 'type'] == 'Goal' and df_team.loc[j, 'teamName']==team_name:
                    ax.scatter(df_team.loc[i, 'y'],df_team.loc[i, 'x'], s=750, marker='*', color='#39e600', edgecolor='k', zorder=3)
                    goal_count += 1
                    break
                j += 1

    shot_count = 0
    # Iterate through the DataFrame
    for i in range(len(df_team)):
        if ((df_team.loc[i, 'type'] in ['BallRecovery', 'Interception']) and 
            (df_team.loc[i, 'teamName'] == team_name) and 
            (df_team.loc[i, 'x'] >= 63)):
            
            possession_id = df_team.loc[i, 'possession_id']
            
            # Check the following rows within the same possession
            j = i + 1
            while j < len(df_team) and df_team.loc[j, 'possession_id'] == possession_id and df_team.loc[j, 'teamName']==team_name:
                if ('Shot' in df_team.loc[j, 'type']) and (df_team.loc[j, 'teamName']==team_name):
                    ax.scatter(df_team.loc[i, 'y'],df_team.loc[i, 'x'], s=200, color='#a331f5', edgecolor='#f5f5f5', zorder=2)
                    shot_count += 1
                    break
                j += 1
    
    ht_count = 0
    p_list = []
    # Iterate through the DataFrame
    for i in range(len(df_team)):
        if ((df_team.loc[i, 'type'] in ['BallRecovery', 'Interception']) and 
            (df_team.loc[i, 'teamName'] == team_name) and 
            (df_team.loc[i, 'x'] >= 63)):
            
            # Check the following rows
            j = i + 1
            if ((df_team.loc[j, 'teamName']==team_name) and
                (df_team.loc[j, 'type']!='Dispossessed') and (df_team.loc[j, 'type']!='OffsidePass')):
                ax.scatter(df_team.loc[i, 'y'],df_team.loc[i, 'x'], s=150, color='None', edgecolor='#ff4b44')
                ht_count += 1
                p_list.append(df_team.loc[i, 'shortName'])

    ax.text(34, 57, 'Top Ball Winners', ha='center', va='center', fontsize=20, bbox=dict(facecolor=bg_color, edgecolor=bg_color, 
                                                                                         boxstyle='round,pad=0.3'))
    counter = Counter(p_list)
    top_3_shortNames = counter.most_common(3)
    y = 54
    for shortName, count in top_3_shortNames:
        ax.text(34, y, f'{shortName}: {count}', ha='center', va='top', fontsize=17, bbox=dict(facecolor=bg_color, edgecolor=bg_color, 
                                                                                         boxstyle='round,pad=0.3'))
        y = y - 4
    # ax.text(68, 49, f'{top_3_shortNames[0][0]}: {top_3_shortNames[0][1]}', ha='left', va='center', fontsize=15)
    # ax.text(34, 49, f'{top_3_shortNames[1][0]}: {top_3_shortNames[1][1]}', ha='center', va='center', fontsize=15)
    # ax.text(0, 49, f'{top_3_shortNames[2][0]}: {top_3_shortNames[2][1]}', ha='right', va='center', fontsize=15)
                
    
    ax.hlines(63,xmin=-2,xmax=70, color='gray', linestyle='--', linewidth=2)
    ax.text(60, 117, f'{team_name}', fontsize=35, fontweight='bold', ha='left', va='center')
    arrow = patches.FancyArrowPatch((70,86), (70, 105), arrowstyle='->',
                                        color='gray', mutation_scale=15, linewidth=1.5)
    ax.add_patch(arrow)
    arrow = patches.FancyArrowPatch((70,82), (70, 63), arrowstyle='->',
                                        color='gray', mutation_scale=15, linewidth=1.5)
    ax.add_patch(arrow)
    ax.text(69,84.5, '40%', fontsize=13, fontweight='bold', color='gray', ha='right', va='center')
    ax.text(69,83, 'of length', fontsize=6, fontweight='bold', color='gray', ha='right', va='center')
    # ax.text(66, 59.5, f'High Turnovers: {ht_count}', color='#ff4b44', fontsize=14, fontweight='bold')
    # ax.text(66, 56.75, f'Shot-ending: {shot_count}', color='#a331f5', fontsize=14, fontweight='bold')
    # ax.text(66, 54, f'Goal-ending: {goal_count}', color='#33cc00', fontsize=14, fontweight='bold')
    ax_text(60, 112, f'<High Turnovers: {ht_count}> | <Shot-ending: {shot_count}> | <Goal-ending: {goal_count}>', fontsize=15,
            highlight_textprops=[{'color':'#ff4b44'}, {'color':'#a331f5'}, {'color':'#33cc00'}], ax=ax, ha='left', va='center')
    ax.text(60, 109, f'Data from: Opta | Till GW10 in LaLiga 2024-25 | made by: @adnaaan433', fontsize=13, color='k', ha='left', va='center')

    IMAGE_URL = f'file:///D:/FData/club_teams_logo/all_league%20-%20Copy/{team_name}.html'
    image = Image.open(urlopen(IMAGE_URL))
    ax_image = pitch.inset_image(113, 68, image, height=10, ax=ax)
    
    return

fig,ax=plt.subplots(figsize=(10,10), facecolor='#f5f5f5')
plot_highTurnover(ax, 'Real Madrid')
```

```python
# List of team names
team_names = df.teamName.unique()

# Create a DataFrame with team names and their ht_count values
team_ht = pd.DataFrame({'teamName': team_names})
team_ht['ht_count'] = team_ht['teamName'].apply(team_ht_list)

# Sort the teams by ht_count value
team_ht_sorted = team_ht.sort_values(by='ht_count', ascending=False)

top12 = team_ht_sorted.head(12)
bot10 = team_ht_sorted.tail(10)

# Create the figure and axes for a 4x5 grid
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(40, 27), facecolor=bg_color)

# Flatten the axes array to easily loop through it
axes = axes.ravel()

# Loop through the sorted team names and corresponding axes
for i, row in enumerate(top12.itertuples()):
    team_name = row.teamName
    # ht_count = row.ht_count
    plot_highTurnover(axes[i], team_name)

fig.text(0.22, 1.025, 'LaLiga Teams High-Turnovers Analysis', fontsize=60, fontweight='bold', ha='left', va='top', fontproperties=custom_font)
fig.text(0.22, 0.985, "Teams are sorted by Total Number of High-Turnovers\nHigh-Turnover means Possession won in the final 40% lenght of the pitch",
         fontsize=35, ha='left', va='top', fontproperties=custom_font_thin)
fig.text(0.22, 0.94, "Data till GW9 of LaLiga 2024-25 | made by: @adnaaan433 | inspired from: @Opta",
         fontsize=35, ha='left', va='top', fontproperties=custom_font_thin)

image_url = 'file:///D:/FData/Leagues_logos/LaLiga%20.html'
image = urlopen(image_url)
image = Image.open(image)
ax_image = add_image(image, fig, left=0.125, bottom=0.93, width=0.09, height=0.09)
```

```python
def plot_highTurnover_against(ax, team_name):
    df_team = df[(df['teamName']==team_name) | (df['oppositionTeamName']==team_name)]
    df_team = df_team.reset_index(drop=True)

    pitch = VerticalPitch(pitch_type='uefa', pitch_color='#f5f5f5', line_color='k', corner_arcs=True)
    pitch.draw(ax=ax)
    ax.set_ylim(52.25, 105.25)
    # ax.set_xlim(70, -0.25)

    goal_count = 0
    # Iterate through the DataFrame
    for i in range(len(df_team)):
        if ((df_team.loc[i, 'type'] in ['BallRecovery', 'Interception']) and 
            (df_team.loc[i, 'oppositionTeamName'] == team_name) and 
            (df_team.loc[i, 'x'] >= 63)):
            
            possession_id = df_team.loc[i, 'possession_id']
            
            # Check the following rows within the same possession
            j = i + 1
            while j < len(df_team) and df_team.loc[j, 'possession_id'] == possession_id and df_team.loc[j, 'oppositionTeamName']==team_name:
                if df_team.loc[j, 'type'] == 'Goal' and df_team.loc[j, 'oppositionTeamName']==team_name:
                    ax.scatter(df_team.loc[i, 'y'],df_team.loc[i, 'x'], s=500, marker='*', color='#39e600', edgecolor='k', zorder=3)
                    goal_count += 1
                    break
                j += 1

    shot_count = 0
    # Iterate through the DataFrame
    for i in range(len(df_team)):
        if ((df_team.loc[i, 'type'] in ['BallRecovery', 'Interception']) and 
            (df_team.loc[i, 'oppositionTeamName'] == team_name) and 
            (df_team.loc[i, 'x'] >= 63)):
            
            possession_id = df_team.loc[i, 'possession_id']
            
            # Check the following rows within the same possession
            j = i + 1
            while j < len(df_team) and df_team.loc[j, 'possession_id'] == possession_id and df_team.loc[j, 'oppositionTeamName']==team_name:
                if ('Shot' in df_team.loc[j, 'type']) and (df_team.loc[j, 'oppositionTeamName']==team_name):
                    ax.scatter(df_team.loc[i, 'y'],df_team.loc[i, 'x'], s=75, color='#a331f5', edgecolor='#f5f5f5', zorder=2)
                    shot_count += 1
                    break
                j += 1
    
    ht_count = 0
    # Iterate through the DataFrame
    for i in range(len(df_team)):
        if ((df_team.loc[i, 'type'] in ['BallRecovery', 'Interception']) and 
            (df_team.loc[i, 'oppositionTeamName'] == team_name) and 
            (df_team.loc[i, 'x'] >= 63)):
            
            # Check the following rows
            j = i + 1
            if ((df_team.loc[j, 'oppositionTeamName']==team_name) and
                (df_team.loc[j, 'type']!='Dispossessed') and (df_team.loc[j, 'type']!='OffsidePass')):
                ax.scatter(df_team.loc[i, 'y'],df_team.loc[i, 'x'], s=55, color='None', edgecolor='#ff4b44')
                ht_count += 1
                

    df_dsps = df_team[(df_team['x']<=42) & (df_team['type']=='Dispossessed') & (df_team['teamName']==team_name)]
    df_usto = df_team[(df_team['x']<=42) & (df_team['type']=='TakeOn') & (df_team['outcomeType']=='Unsuccessful') & (df_team['teamName']==team_name)]
    df_errr = df_team[(df_team['x']<=42) & (df_team['type']=='Error') & (df_team['teamName']==team_name)]
    df_pint = df_team[(df_team['endX']<=42) & (df_team['type']=='Pass') & (df_team['type'].shift(-1)=='Interception') & (df_team['teamName']==team_name)]
    df_ball_losse = pd.concat([df_dsps, df_usto, df_errr, df_pint])
    err_count = df_ball_losse['shortName'].value_counts().head(3)
    for i, (name, count) in enumerate(err_count.items()):
        ax.text(34, 111+(i*3), f'{name}: {count}', fontsize=18, ha='center', va='top')
    
    ax.text(34, 109, 'Most Possession Lost by:', ha='center', va='center', fontsize=18)
    ax.hlines(63,xmin=-2,xmax=70, color='gray', linestyle='--', linewidth=2)
    
    ax.text(60, 40, f'{team_name}', fontsize=35, fontweight='bold', ha='left', va='center')
    
    arrow = patches.FancyArrowPatch((70,86), (70, 105), arrowstyle='->',color='gray', mutation_scale=15, linewidth=1.5)
    ax.add_patch(arrow)
    arrow = patches.FancyArrowPatch((70,82), (70, 63), arrowstyle='->',color='gray', mutation_scale=15, linewidth=1.5)
    ax.add_patch(arrow)
    
    ax.text(69, 83, '40%', fontsize=13, fontweight='bold', color='gray', ha='right', va='top')
    ax.text(69, 85, 'of length', fontsize=6, fontweight='bold', color='gray', ha='right', va='center')
    # ax.text(66, 55, f'High Turnovers: {ht_count}', color='#ff4b44', fontsize=13, fontweight='bold')
    # ax.text(66, 57, f'Shot-ending: {shot_count}', color='#a331f5', fontsize=13, fontweight='bold')
    # ax.text(66, 59, f'Goal-ending: {goal_count}', color='#33cc00', fontsize=13, fontweight='bold')
    ax_text(60, 45, f'<Dangerous Possession Lost: {ht_count}> | <Shot-ending: {shot_count}> | <Goal-ending: {goal_count}>', fontsize=13,
            highlight_textprops=[{'color':'#ff4b44'}, {'color':'#a331f5'}, {'color':'#33cc00'}], ax=ax, ha='left', va='center')
    ax.text(60, 48, f'Data from: Opta | Till GW10 in LaLiga 2024-25 | made by: @adnaaan433', fontsize=12, color='k', ha='left', va='center')

    IMAGE_URL = f'file:///D:/FData/club_teams_logo/all_league%20-%20Copy/{team_name}.html'
    image = Image.open(urlopen(IMAGE_URL))
    ax_image = pitch.inset_image(46, 68, image, height=10, ax=ax)
    
    ax.invert_yaxis()
    return

fig,ax=plt.subplots(figsize=(10,10), facecolor='#f5f5f5')
plot_highTurnover_against(ax, 'Real Madrid')
```
