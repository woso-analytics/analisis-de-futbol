```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.patheffects as path_effects
import matplotlib.font_manager as fm
from highlight_text import fig_text, ax_text
from adjustText import adjust_text
from matplotlib.colors import LinearSegmentedColormap, Normalize
from matplotlib import cm
import scipy.stats as stats


from mplsoccer import Pitch, add_image

from PIL import Image
import urllib
import os
```

```python
df_passing = pd.read_csv('data/fbref_player_passing_wsl.csv', index_col=0)
df_possession = pd.read_csv('data/fbref_player_possession_wsl.csv', index_col=0)
# df_passing = pd.read_csv('data/fbref_player_passing_ligaf.csv', index_col=0)
# df_possession = pd.read_csv('data/fbref_player_possession_ligaf.csv', index_col=0)
```

```python
columns_to_keep = ['Player', 'carries_into_pen_area']

# Drop other columns
columns_to_drop = [col for col in df_possession.columns if col not in columns_to_keep]
df_possession = df_possession.drop(columns=columns_to_drop)
```

```python
# player = df_possession['Player'].value_counts().get('Chloe Kelly', 0)
# player
df_possession1 = df_possession[df_possession['Player'].value_counts() == 1]
```

```python
df_possession['Player'].value_counts() ==1
```

```python
df = pd.merge(df_passing, df_possession, how = 'left', on = [ 'Player', 'Player'])
```

```python
df
```

```python
df['total_passes_into_pen_area'] = df['crosses_into_pen_area'] + df['passes_into_pen_area']
```

```python
df['player_first_name']= df['Player'].apply(lambda x: x.split()[0])

# Function to extract last names
def get_last_name(full_name):
    if pd.isna(full_name):
        return full_name
    parts = full_name.split()
    if len(parts) == 1:
        return None # No need for last name if there's only one word
    elif len(parts) == 2:
        return parts[1]
    else:
        return parts[1][0] + ". " + " ".join(parts[2:])

# Applying the function to create 'shortName' column
df['player_last_name'] = df['Player'].apply(get_last_name)
df['90s'] = df['90s'].astype(float)
df['minutes'] = df['90s'] * 90
```

```python
df = (
    df.groupby(['player_first_name', 'player_last_name'])
    [['minutes', 'total_passes_into_pen_area', 'carries_into_pen_area']].sum()
    .reset_index()
)
df
```

```python
df_plot = df[(df['minutes'] > 675) &(df['total_passes_into_pen_area'] >= df['total_passes_into_pen_area'].median()) & (df['carries_into_pen_area'] >= df['carries_into_pen_area'].median())]
df_plot
```

```python
df_plot = df_plot[df_plot['player_last_name']!='Kelly']
```

```python
df_plot = df_plot.assign(passes_per_90 = lambda x: (x.total_passes_into_pen_area/x.minutes)*90)
df_plot = df_plot.assign(carries_per_90 = lambda x: (x.carries_into_pen_area/x.minutes)*90)
```

```python
df_plot.sort_values(by='passes_per_90')
```

```python
df_plot['zscore'] = stats.zscore(df_plot['passes_per_90'])*.4 + stats.zscore(df_plot['carries_per_90'])*.6
df_plot['annotated'] = [True if x > df_plot['zscore'].quantile(.8) else False for x in df_plot['zscore']]
```

```python
fig = plt.figure(figsize=(8,8), dpi=100)
ax = plt.subplot()
ax.grid(visible=True, ls='--', color='lightgrey')

ax.scatter(
    df_plot['passes_per_90'], df_plot['carries_per_90'], 
    c=df_plot['zscore'], cmap='inferno', 
    zorder=3, ec='grey', s=55, alpha=0.8)
    
texts = []
annotated_df = df_plot[df_plot['annotated']].reset_index(drop=True)
for index in range(annotated_df.shape[0]):
    texts += [
        ax.text(
            x=annotated_df['passes_per_90'].iloc[index], y=annotated_df['carries_per_90'].iloc[index],
            s=f"{annotated_df['player_first_name'].iloc[index][0]}. {annotated_df['player_last_name'].iloc[index]}",
            path_effects=[path_effects.Stroke(linewidth=2, foreground=fig.get_facecolor()), 
            path_effects.Normal()], color='black',
            family='DM Sans', weight='bold'
        )
    ]

adjust_text(texts, only_move={'points':'y', 'text':'xy', 'objects':'xy'})

# ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
# ax.yaxis.set_major_locator(ticker.MultipleLocator(.1))
# ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:.0%}'))
# ax.set_xlim(0)
# ax.set_ylim(0,1)

ax.set_ylabel('Carries into penalty area per 90')
ax.set_xlabel('Passes into penalty area per 90')

fig_text(
    x = 0.09, y = .99, 
    s = "Barclays WSL: Ball in the box",
    va = "bottom", ha = "left",
    fontsize = 20, color = "black", font = "DM Sans", weight = "bold"
)

fig_text(
    x = 0.09, y = 0.91, 
    s = "Passes and carries into penalty area per 90.\nOnly players with above median total passes & median total carries are shown.\nSeason 2024/2025 | Data from fbref | Only players with more than 7.5 90s",
    va = "bottom", ha = "left",
    fontsize = 12, color = "#5A5A5A", font = "DM Sans"
)

fotmob_url = 'https://images.fotmob.com/image_resources/logo/leaguelogo/9227.png'
# fotmob_url ='https://images.fotmob.com/image_resources/logo/leaguelogo/9907.png'
club_icon = Image.open(urllib.request.urlopen(f'{fotmob_url}'))
ax_timage = add_image(club_icon, fig, left=0.84, bottom=0.98, width=0.050, height=0.050)

fig_text(
    x = 0.75, y = 0.06,
    s = "@woso_analytics",
    color = '#1C2C5B',
    size = 7,
    annotationbbox_kw={"xycoords": "figure fraction"}
)

plt.savefig(
	"figures/16052024_pen_area_wsl_with_carries.png",
	dpi = 600,
	facecolor = "#EFE9E6",
	bbox_inches="tight",
    edgecolor="none",
	transparent = False
)

plt.savefig(
	"figures/16052024_pen_area_wsl_with_carries_tr.png",
	dpi = 600,
	facecolor = "none",
	bbox_inches="tight",
    edgecolor="none",
	transparent = True
)
```
