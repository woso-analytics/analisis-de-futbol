```python
from mplsoccer.pitch import Pitch, VerticalPitch
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
from highlight_text import HighlightText, ax_text, fig_text
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
from scipy.ndimage import gaussian_filter
import matplotlib.patheffects as path_effects
import glob 
import os
```

```python
path = r'C:\Users\Hp2\Music\joel\An√°lisis de fulbo\WSL24-25' # use your path
all_files = glob.glob(os.path.join(path, "*.csv"))

data = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)

data['x'] = data['x']/1.05
data['y'] = data['y']/0.68
data['endX'] = data['endX']/1.05
data['endY'] = data['endY']/0.68
data['goalMouthY'] = data['goalMouthY']/0.68
```

```python
data['x'] = data['x']*1.2
data['y'] = data['y']*.8
data['endX'] = data['endX']*1.2
data['endY'] = data['endY']*.8
data['goalMouthY'] = data['goalMouthY']/.8

green = '#69f900'
red = '#cb2217'
blue = '#00285e'
sky_blue = '#6caddf'
bg_color= '#fffcf7'
gold = '#ffc758'
line_color= '#000000'
bbox_pad = .8
bboxprops = {'linewidth': 0, 'pad': bbox_pad}
```

```python
data_carries = data[data['type'] == 'Carry']
data_passes = data[(data['type'] == 'Pass') & (data['outcomeType'] == 'Successful')]
```

```python
# Define the Zone 14 coordinates
z14_x_min = 78
z14_x_max = 78 + 24
z14_y_min = 30
z14_y_max = 30 + 20



# Function to identify actions and calculate progressive actions starting inside the specified zone
def calculate_progressive_actions(df, x_min, x_max, y_min, y_max):
    df_prog = df.copy()
    df_prog['in_zone'] = (df_prog['x'].between(x_min, x_max) & df_prog['y'].between(y_min, y_max))
    df_prog['beginning'] = np.sqrt(np.square(120 - df_prog['x']) + np.square(40 - df_prog['y']))
    df_prog['end'] = np.sqrt(np.square(120 - df_prog['endX']) + np.square(40 - df_prog['endY']))
    df_prog['progressive'] = (df_prog['end'] / df_prog['beginning']) < 0.75
    return df_prog[df_prog['progressive'] & df_prog['in_zone']]

# Calculate progressive actions out of Zone 14
prog_z14_passes = calculate_progressive_actions(data_passes, z14_x_min, z14_x_max, z14_y_min, z14_y_max)
prog_z14_carries = calculate_progressive_actions(data_carries, z14_x_min, z14_x_max, z14_y_min, z14_y_max)

# Group and aggregate
z14_passes_grouped = prog_z14_passes.groupby(['playerId', 'name', 'teamName']).size().reset_index(name='z14_prog_passes')
z14_carries_grouped = prog_z14_carries.groupby(['playerId', 'name', 'teamName']).size().reset_index(name='z14_prog_carries')

# Merge and combine
combined_z14_prog_df = pd.merge(z14_passes_grouped, z14_carries_grouped, on=['playerId', 'name', 'teamName'], how='outer').fillna(0)

# Calculate total progressive actions and per 90 numbers
combined_z14_prog_df['total_prog_actions'] = combined_z14_prog_df['z14_prog_passes'] + combined_z14_prog_df['z14_prog_carries']

# Merge with player metadata (e.g., minutes played)
info = pd.read_csv("minutes.csv")
# info['90s'] = info['Mins'] / 90

# Merge with the main dataframe
combined_z14_prog_df = pd.merge(combined_z14_prog_df, info, on=['name'], how='left')
combined_z14_prog_df = combined_z14_prog_df[(combined_z14_prog_df['90s'] >= 12) & (combined_z14_prog_df['position'] != 'GK')
                                            ]

# Calculate per 90 values
combined_z14_prog_df['z14_prog_passes_p90'] = combined_z14_prog_df['z14_prog_passes'] / combined_z14_prog_df['90s']
combined_z14_prog_df['z14_prog_carries_p90'] = combined_z14_prog_df['z14_prog_carries'] / combined_z14_prog_df['90s']
combined_z14_prog_df['total_prog_actions_p90'] = combined_z14_prog_df['total_prog_actions'] / combined_z14_prog_df['90s']

# Get the top 12 players based on per 90 numbers
top_12_prog_z14 = combined_z14_prog_df.sort_values("total_prog_actions_p90", ascending=False).head(30)

top_12_prog_z14
```

```python
fig, axs = plt.subplots(3, 4, figsize=(30, 18))
fig.set_facecolor(bg_color)

# Define the title and additional text
fig_text(0.45, 1.00, "Barclays WSL 24-25 - Top 12 Players for Zone 14 Progressive Actions per 90", ha="center", font='DM Sans', color="k", fontsize=36, fontweight='bold')
fig_text(0.278, 0.97, "<Progressive Carries> and <Progressive Passes>", ha="center", font='DM Sans', color="k", fontsize=30, fontweight='bold', highlight_textprops=[{"color": red}, {"color": blue}])
fig_text(0.299, 0.94, "Outfield Players Only | Players with at least 12 90s Played | Corners Included", ha="center", font='Arial Rounded MT Bold', color='k', fontsize=20, fontweight='bold')

for ax, (_, player_data) in zip(axs.flatten(), top_12_prog_z14.iterrows()):
    player_id = player_data['playerId']
    player_name = player_data['name']
    
    player_prog_z14_passes = prog_z14_passes[prog_z14_passes['playerId'] == player_id]
    player_prog_z14_carries = prog_z14_carries[prog_z14_carries['playerId'] == player_id]
    
    pitch = Pitch(pitch_type='statsbomb', pitch_color=bg_color, line_color='k')
    pitch.draw(ax=ax)
    
    # Plot Zone 14 Progressive Passes
    pitch.lines(player_prog_z14_passes.x, player_prog_z14_passes.y, player_prog_z14_passes.endX, player_prog_z14_passes.endY,
                lw=3, transparent=True, comet=True, color=blue, ax=ax)
    pitch.scatter(player_prog_z14_passes.endX, player_prog_z14_passes.endY, s=40, marker='o', edgecolors='none', c=blue, ax=ax, alpha=1)
    
    # Plot Zone 14 Progressive Carries
    pitch.lines(player_prog_z14_carries.x, player_prog_z14_carries.y, player_prog_z14_carries.endX, player_prog_z14_carries.endY, 
                ls='dashed', lw=2, transparent=False, comet=False, color=red, ax=ax)
    pitch.scatter(player_prog_z14_carries.endX, player_prog_z14_carries.endY, s=40, marker='o', edgecolors='none', c=red, ax=ax, alpha=1)
    
    ax.invert_yaxis()
    ax.set_title(f'{player_name}', font='DM Sans', color='k', fontsize=16)
    
    ax.text(5, 5, f'Z14 Carries: {player_data["z14_prog_carries"]}', color=red, font='Arial Rounded MT Bold', fontsize=12, ha='left')
    ax.text(5, 13, f'Z14 Passes: {player_data["z14_prog_passes"]}', font='Arial Rounded MT Bold', color=blue, fontsize=12, ha='left')
    ax.text(28, -7, f'Z14 Prog. Actions p90: {player_data["total_prog_actions_p90"]:.2f}', font='Arial Rounded MT Bold', color='k', fontsize=14, ha='left')

    
    
fig_text(0.18, 0.06, "Inspired by: @pranav_m28", ha="center", font='DM Sans' ,color="k", fontsize=20, fontweight='bold')
fig_text(0.86, 0.06, "Data from Opta", ha="center", font='DM Sans' ,color="k", fontsize=20, fontweight='bold')


fig_text(
    0.510, 0.070, "Direction of Play: Left to Right", font='DM Sans' , size=18,
    ha="center", color="k" , fontweight='bold'
)

plt.savefig("Z14_Prog_Out_PL.png" , dpi=300, bbox_inches='tight')    
    
plt.show()
```
