```python
def getAttackingSpeed(team, df):
    """
        Speed is the distance covered divided by time taken

        This function is to get the average speed of a team. It does so by calculating
        the speed of every possession sequence i.e. get the possession_id of each 
        possession sequence by a team and calculate the distance covered and the time taken 
        to do so and then average the results.

            Args: team (string) the name of the team we wish to find the average attacking speed of
                  df (pd.Dataframe) the dataframe of Opta event data with the events we wish to 
                  calculate the attacking speed from

            Returns: speed (float) the average attacking speed of the team in m/s            
    """
    # Make a list to store the speed of every sequence
    attacking_speed_list = []
    # Iterate through the dataframe to find the last occurence of possession_id i.e the end of the possession sequence
    for i in range(len(df)):
        # Get initial possession_id
        possession_id = df.loc[i, 'possession_id']
        # Find the last point with the possession_id i.e last point in possession sequence
        # Check the following rows within the same possession
        j = i + 1
        while j < len(df) and df.loc[j, 'possession_id'] == possession_id and df.loc[j, 'possession_team']==team:
            j += 1
            # Calculate the distance and time between the first and last point in the possession sequence and hence the speed
        if df.loc[j ,'x'] >= df.loc[i ,'x']:
            distance = ((df.loc[j ,'x'] - df.loc[i ,'x'])**2 + (df.loc[j, 'y'] - df.loc[i, 'y'])**2)**0.5
        if df.loc[j ,'x'] <= df.loc[i ,'x']:
            distance = -((df.loc[j ,'x'] - df.loc[i ,'x'])**2 + (df.loc[j, 'y'] - df.loc[i, 'y'])**2)**0.5
        time = df.loc[j, 'minute'] * 60 + df.loc[j, 'second'] - df.loc[i, 'minute'] * 60 + df.loc[i, 'second']
        if time > 0:
            speed = distance/time
            attacking_speed_list.append(speed)
        # Move to the next possession_id / possession sequence by our team
        while df.loc[j, 'possession_team'] != team:
            j+=1
        i = j + 1

    return np.series(attacking_speed_list).mean()
    
city = getAttackingSpeed("Manchester City Women", df)
city
```

```python
def ppda(teamName):
    # PPDA
#     if len(def_acts) == 0
    def_acts = df[(df['teamName']==teamName) & (df['type'].str.contains('Interception|Foul|Challenge|BlockedPass|Tackle')) & (df['x']>35)]
    opposition_passes = df[(df['oppositionTeamName']==teamName) & (df['type']=='Pass') & (df['outcomeType']=='Successful') & (df['x']<70)]
    ppda = round((len(opposition_passes)/len(def_acts)), 2)
    
    return ppda

ppda("Crystal Palace FC Women")
# ppda("West Ham United Women")
```
