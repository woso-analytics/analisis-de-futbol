```python
def plot_kdetouches(ax, fig, playerId, countryId, playerName, df=data):
    '''
    Plot kde touches for a specific player.
    '''
    data = df.copy()
    data_player = data[data['playerId'] == int(playerId)].reset_index(drop=True)
    total_touches = data_player.shape[0]
    touches_third = data_player[data_player['x'] >= (100*2)/3].shape[0]/total_touches
    pitch = VerticalPitch(
        pitch_type='opta',
        goal_type='box',
        linewidth=1.25,
        line_color='black',
        pitch_color='#F6F7F8'
    )
    pitch.draw(ax=ax)

    pitch.kdeplot(
        data_player.x, data_player.y, 
        ax=ax, fill=True, 
        levels=100, shade_lowest=True,
        cut=4, zorder=-1, cmap='SOC1')

    ax.scatter(data_player.y, data_player.x, s=10, alpha=0.2, lw=1.5, color='#0C2C32')
    # -- Transformation functions
    ax.set_ylim(ax.get_ylim()[0], ax.get_ylim()[1])
    ax.set_xlim(ax.get_xlim()[0], ax.get_xlim()[1])
    DC_to_FC = ax.transData.transform
    FC_to_NFC = fig.transFigure.inverted().transform
    # -- Take data coordinates and transform them to normalized figure coordinates
    DC_to_NFC = lambda x: FC_to_NFC(DC_to_FC(x))
    ax_coords = DC_to_NFC((15,109))
    ax_size = 0.018
    image_ax = fig.add_axes(
        [ax_coords[0], ax_coords[1], ax_size, ax_size],
        fc='None', anchor='C'
    )
    fotmob_url = 'https://images.fotmob.com/image_resources/logo/teamlogo/'
    player_face = Image.open(urllib.request.urlopen(f"{fotmob_url}{countryId}.png"))
    image_ax.imshow(player_face)
    image_ax.axis("off")
    ax_text(
        x=95, y=115,
        s=f'{playerName}\n<Touches {total_touches:.0f} | Final Third: {touches_third:.0%}>',
        highlight_textprops=[{'size':7, 'weight':'normal'}],
        ax=ax,
        weight='bold',
        size=10,
        ha='left'
    )
    return ax
```

```python
layout_ = '''
    ABC
    DEF
'''

fig = plt.figure(figsize=(8,8), dpi=300, facecolor='#F6F7F8')
axs = fig.subplot_mosaic(
    layout_
)

counter = 0
plt.rcParams["font.family"] = "DM Sans"

for k, ax in axs.items():
    playerId = players[counter]['whoscored_id']
    countryId = players[counter]['country_id']
    playerName = players[counter]['name']
    plot_kdetouches(ax, fig, playerId, countryId, playerName)
    counter += 1

axs['A'].annotate(
    xy=(93, 5),
    text='Attacking direction',
    rotation=90,
    size=7
)

axs['A'].annotate(
    xy=(85,2),
    xytext=(85,45),
    text='',
    arrowprops=dict(arrowstyle='<-',linewidth=0.55)
)

fig_text(
    x = 0.20, y = .96, 
    s = "Manchester City Women Touch Maps Brighton",
    va = "bottom", ha = "left",
    fontsize = 14, color = "black", weight = "bold"
)

fig_text(
    x = 0.83, y = .96, 
    s = "| <Inside MCWFC>",
    va = "bottom", ha = "left",
     highlight_textprops = [
        {"color": "#1C2C5B", "weight": "bold"}
    ],
    color = "#D4A12A",
    fontsize = 9
)

fig_text(
	x = 0.20, y = .94, 
    s = "Touches per player in the fixture against Brighton & Hove Albion Women.",
	va = "bottom", ha = "left",
	fontsize = 7, color = "#4E616C"
)

# fig_text(
#     x = 0.87, y = 0.15,
#     s = "Viz by: @joellikesthis| Inspired by @sonofacorner.",
#     color = "black",
#     size = 4,
#     annotationbbox_kw={"xycoords": "figure fraction"}
# )

plt.savefig(
	"09052022_brighton_touches.png",
	dpi = 600,
	facecolor = "#F6F7F8",
	bbox_inches="tight",
    edgecolor="none",
	transparent = False
)

plt.savefig(
	"09052022_brighton_touches_tr.png",
	dpi = 600,
	facecolor = "none",
	bbox_inches="tight",
    edgecolor="none",
	transparent = True
)

city_url = 'https://images.fotmob.com/image_resources/logo/teamlogo/231488.png'
club_icon = Image.open(urllib.request.urlopen(f'{city_url}'))
# image_ax.imshow(club_icon)
# image_ax.axis('off')
ax_timage = add_image(club_icon, fig, left=0.13, bottom=0.94, width=0.050, height=0.050)
```
